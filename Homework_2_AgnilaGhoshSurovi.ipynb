As an illustration, I include a short example where I used Python to compute the **analytical stress relaxation and creep response of viscoelastic soft materials**.  
These analytical solutions were employed to validate my finite element results for viscoelastic models with different mechanical parameters.

````python
# -*- coding: utf-8 -*-
"""
Created on Mon May 19 16:54:38 2025

@author: Agnila
"""

import numpy as np
import matplotlib.pyplot as plt

# Time array
t = np.linspace(0, 160, 1000)

# Constant strain and stress for simulations
epsilon_0 = 0.1  # strain
sigma_0 = 0.1      # constant stress

# Material models (E1, E2, eta)
material_models = [
    (5, 20, 10),
    (5, 20, 40),
    (5, 20, 100),
    (10, 40, 20),
    (10, 40, 80),
    (10, 40, 200),
    (30, 120, 60),
    (30, 120, 240),
    (30, 120, 600)
]

# Initialize plots
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

# Plot each material model
for i, (E1, E2, eta) in enumerate(material_models, 1):
    # Relaxation time
    tau_R = eta / E2
    # Creep time
    tau_C = eta * (E1 + E2) / (E1 * E2)

    # Stress relaxation: σ(t) = [σ0 - E1*ε0]*exp(-t/τR) + E1*ε0
    # sigma_t = (sigma_0 - (E1*epsilon_0))*np.exp(-t / tau_R) + E1* epsilon_0
    # Stress relaxation: σ(t) = [E1 + E2 * exp(-t/τR)] * ε0
    sigma_t = (E1 + E2 * np.exp(-t / tau_R)) * epsilon_0

    # Creep: ε(t) = (σ0 / E1) * [1 - (E2 / (E1 + E2)) * exp(-t/τC)]
    epsilon_t = (sigma_0 / E1) * (1 - (E2 / (E1 + E2)) * np.exp(-t / tau_C))

    # Plot
    ax1.plot(t, sigma_t, label=f'Model {i}')
    ax2.plot(t, epsilon_t, label=f'Model {i}')

# Format stress relaxation plot
ax1.set_title("Stress Relaxation Behavior")
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Stress σ(t) [kPa]")
ax1.legend()
ax1.grid(True)

# Format creep plot
ax2.set_title("Creep Behavior")
ax2.set_xlabel("Time (s)")
ax2.set_ylabel("Strain ε(t)")
ax2.legend()
ax2.grid(True)

# Save and show
fig1.savefig("stress_relaxation.png", dpi=300)
fig2.savefig("creep_behavior.png", dpi=300)
plt.show()
````

The stress relaxation plots show that σ(t) decreases from (E1+E2)·ε0 to E1·ε0 over time, while creep curves increase from σ0/(E1+E2) to σ0/E1.
