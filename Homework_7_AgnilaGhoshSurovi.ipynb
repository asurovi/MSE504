#### Define the Casino with the 3 Bandits with the probabilities of win being given by Gaussian distributions with the parameters (mean, dispersion) being (6, 2) for the first one (A), (5,3) for the second one (B), and (5.5, 0.2) for the third one (C). You can adapt the code from the example notebook.

Bandits:
- A: mean=6,   stdev=2
- B: mean=5,   stdev=3
- C: mean=5.5, stdev=0.2

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from dataclasses import dataclass
from typing import List, Tuple, Dict
```

```python
# random number generator (RNG) with a fixed seed = 3031
rng = np.random.default_rng(3031)

# display function for table display
def display_dataframe_to_user(title, df):
    print(f"\n--- {title} ---")
    display(df)

# Defines one “arm” that produces values drawn from a Gaussian distribution.
# mean and stdev (standard deviation) are stored for later.
# Draws a random number from N(mean, stdev) using rng.normal.
# Rounds to one decimal (to mimic how the original notebook handled samples).
# Returns it as a float.
class GaussianBandit:
    def __init__(self, mean = 0, stdev = 1):
        self.mean = mean
        self.stdev = stdev
    def pull_lever(self):
        return float(np.round(rng.normal(self.mean, self.stdev), 1))

# Converts the Gaussian sample x into a probability between 0 and 1.
# Divides by 10 (so 10 ≈ 100 %), then clips anything below 0 or above 1.
# Example: x = 6 → p = 0.6, x = 12 → p = 1.0 (after clipping).
def prob_from_gaussian_sample(x, scale = 10.0):
    return float(np.clip(x/scale, 0.0, 1.0))

# A dataclass wrapping a GaussianBandit (gb) and a scaling factor.
# It represents a bandit that actually gives binary rewards (win = 1, lose = 0).
# Draw a Gaussian sample x.
# Convert it to a probability p.
# Generate a random number ∈ [0, 1).
# Return 1 if the random number < p (win), otherwise 0 (loss).
@dataclass
class WinProbBandit:
    def __init__(self, name, gb, scale=10.0):
        self.name = name
        self.gb = gb
        self.scale = scale
    def pull(self):
        x = self.gb.pull_lever()
        p = prob_from_gaussian_sample(x, self.scale)
        if rng.random() < p:
          return 1
        else:
          return 0

# Builds a casino containing 3 bandits (A, B and C)
# Each bandit is wrapped inside a WinProbBandit.
class Casino:
    def __init__(self):
        self.names = ["A","B","C"]
        self.bandits = {
            "A": WinProbBandit("A", GaussianBandit(6.0, 2.0)),
            "B": WinProbBandit("B", GaussianBandit(5.0, 3.0)),
            "C": WinProbBandit("C", GaussianBandit(5.5, 0.2)),
        }
    # pull a specific arm by name (“A”, “B”, or “C”).
    def pull(self, name):
      return self.bandits[name].pull()
```

#### Now let's assume that we playing these in parallel. For a single run, calculate the average and dispecrion of cumulative reward as a number of plays. Plot the mean with error bars (dispersion) on the log scale for time. At which point can we decide which of the bandits gives biggest wins?

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 1) Random seed for reproducibility
# -----------------------------
rng = np.random.default_rng(3031)

# -----------------------------
# 2) Bandit & Casino definitions
#    (Gaussian sample -> map to probability -> Bernoulli win)
# -----------------------------
class GaussianBandit:
    def __init__(self, mean, stdev):
        self.mean = float(mean)
        self.stdev = float(stdev)
    def pull_lever(self):
        # draw N(mean, stdev) and round to 1 decimal (like the example notebook)
        return float(np.round(rng.normal(self.mean, self.stdev), 1))

def prob_from_gaussian_sample(x, scale=10.0):
    # map real x to [0,1] probability; e.g., x=6 -> 0.6; clip bounds
    return float(np.clip(x/scale, 0.0, 1.0))

class WinProbBandit:
    def __init__(self, name, gb, scale=10.0):
        self.name = name
        self.gb = gb
        self.scale = scale
    def pull(self):
        x = self.gb.pull_lever()
        p = prob_from_gaussian_sample(x, self.scale)
        return 1 if rng.random() < p else 0

class Casino:
    def __init__(self):
        self.names = ["A","B","C"]
        self.bandits = {
            "A": WinProbBandit("A", GaussianBandit(6.0, 2.0)),
            "B": WinProbBandit("B", GaussianBandit(5.0, 3.0)),
            "C": WinProbBandit("C", GaussianBandit(5.5, 0.2)),
        }
    def pull(self, name):
        return self.bandits[name].pull()

# -----------------------------
# 3) Parallel play (one run)
#    Returns:
#      cum[k]    : cumulative reward (total wins) up to each t
#      cum_std[k]: dispersion (std) of cumulative reward up to each t (plug-in estimate)
# -----------------------------
def parallel_play_one_run(T=5000):
    cas = Casino()
    names = cas.names
    wins = {k: np.zeros(T, dtype=int) for k in names}
    cum  = {k: np.zeros(T, dtype=float) for k in names}
    p_hat = {k: np.zeros(T, dtype=float) for k in names}    # running mean win rate
    cum_std = {k: np.zeros(T, dtype=float) for k in names}  # std of cumulative reward

    for t in range(T):
        n = t + 1
        for k in names:
            r = cas.pull(k)
            wins[k][t] = r
            cum[k][t] = cum[k][t-1] + r if t > 0 else r
            p_hat[k][t] = cum[k][t] / n
            # Binomial plug-in: R_n ~ Binomial(n, p_hat); std(R_n) = sqrt(n p_hat (1 - p_hat))
            cum_std[k][t] = np.sqrt(n * p_hat[k][t] * (1.0 - p_hat[k][t]))
    return cum, cum_std, p_hat

# -----------------------------
# 4) Earliest confident decision
#    (2-sigma rule: top mean - 2*std > others' mean + 2*std)
# -----------------------------
def first_clear_separation(cum, cum_std, names=("A","B","C"), z=2.0):
    T = len(next(iter(cum.values())))
    for t in range(T):
        means = {k: cum[k][t] for k in names}
        stds  = {k: cum_std[k][t] for k in names}
        top = max(names, key=lambda k: means[k])
        ok = True
        for other in names:
            if other == top:
                continue
            # non-overlap condition at t:
            # mean_top - z*std_top > mean_other + z*std_other
            if means[top] - z*stds[top] <= means[other] + z*stds[other]:
                ok = False
                break
        if ok:
            return (t+1), top  # +1 because t is 0-based
    return None, None

# -----------------------------
# 5) Run, plot, and print decision time
# -----------------------------
cum, cum_std, p_hat = parallel_play_one_run(T=5000)

plt.figure(figsize=(8,5))
colors = {"A":"tab:blue", "B":"tab:green", "C":"tab:red"}
for k in ["A","B","C"]:
    t = np.arange(1, len(cum[k])+1)
    # thin error bars to log-spaced checkpoints for readability
    idx = np.unique(np.round(np.logspace(0, np.log10(len(t)), 25)).astype(int)) - 1
    idx = idx[(idx >= 0) & (idx < len(t))]

    plt.plot(t, cum[k], color=colors[k], label=f"{k} cumulative mean")
    plt.errorbar(t[idx], cum[k][idx], yerr=cum_std[k][idx],
                 fmt='o', ms=3, capsize=3, color=colors[k], alpha=0.75)

plt.xscale("log")
plt.xlabel("Time steps = number of plays (log scale)")
plt.ylabel("Cumulative reward (mean) ± std")
plt.title("Parallel play of bandits A, B, C: cumulative reward with dispersion")
plt.legend()
plt.tight_layout()
plt.show()

t_star, winner = first_clear_separation(cum, cum_std)
if t_star is None:
    print("No clear 2σ separation found within the horizon.")
else:
    print(f"Earliest confident decision (2σ non-overlap): at ~{t_star} plays per bandit; winner = {winner}")

# (Optional) also print a few checkpoints
checkpoints = [10, 100, 1000, 5000]
for t in checkpoints:
    print(f"\nAfter {t} plays per bandit:")
    for k in ["A","B","C"]:
        mean_cum = cum[k][t-1]
        std_cum  = cum_std[k][t-1]
        print(f"  Bandit {k}: cumulative mean = {mean_cum:.1f},  dispersion (std) = {std_cum:.1f}")

```

When playing the three Gaussian bandits in parallel,
* The cumulative reward curves separate visibly after roughly 502 plays per bandit.
* Beyond that point, the ranking stabilizes and Bandit A shows the largest average cumulative reward.
* The dispersion (standard deviation) increases steadily with time. The error bars increase with time because we plot cumulative rewards. As more plays are made, both the total reward and its variance grow. This reflects accumulating randomness, not growing uncertainty in the average performance.

#### Now, let's assume that we **do not** know the probability distirbutions of wins, and have to decide which of the hands A, B, or C to play. First, realize the A/B testing strategy. Show the returns for 10, 30, 100, 300, 1000, 3000 test runs and 10000 total number of attempts

```python
# Repeat any policy multiple times
# policy_fn = a strategy (A/B testing, ε-greedy, UCB).
# Creates a new casino each run (so the randomness resets).
# Runs the strategy n_trials times and collects the total rewards.
# Returns the mean and standard deviation of total rewards —
# this is the “Monte Carlo average”.
def simulate_policy(policy_fn, total_attempts: int, n_trials: int = 5):
    totals = []
    for _ in range(n_trials):
        cas = Casino()
        totals.append(policy_fn(cas, total_attempts))
    arr = np.array(totals, dtype=float)
    return float(arr.mean()), float(arr.std())

# A/B testing
def policy_ab(cas: Casino, total_attempts: int, test_runs_per_arm: int) -> int:
    # Exploration phase
    names = cas.names
    test_total = test_runs_per_arm * len(names)
    counts = {k:0 for k in names}
    rewards = {k:0 for k in names}

    # Pull each arm equally test_runs_per_arm times.
    # Record the number of pulls (counts) and total wins (rewards).
    for i in range(test_total):
        k = names[i % len(names)]
        r = cas.pull(k)
        counts[k]+=1
        rewards[k]+=r

    # Compute each arm’s average win rate and pick the best.
    best = max(names, key=lambda k: rewards[k]/max(counts[k],1))

    # Exploitation phase
    remaining = total_attempts - test_total
    total_reward = sum(rewards.values())
    for _ in range(remaining):
      total_reward += cas.pull(best)
    return total_reward

# Each strategy will be run for 10 000 pulls per simulation.
total_attempts = 10_000

# -------------------------------------A/B--------------------------------------
# For each test budget tr, run simulate_policy 5 times.
# Record mean & std of total reward.
# Create a DataFrame (df_ab) and display it.
test_grid = [10, 30, 100, 300, 1000, 3000]
df_ab = pd.DataFrame([
    {"Test runs per arm": tr, "Total attempts": total_attempts, **dict(zip(["Return mean (5 trials)","Return std"],
    simulate_policy(lambda cas, T: policy_ab(cas, T, tr), total_attempts, n_trials=5)))} for tr in test_grid
])
display_dataframe_to_user("A/B testing — returns vs test runs (mean±std over 5 trials)", df_ab)
```

When the underlying win probabilities are unknown, we use an A/B testing policy: first play each arm equally for a fixed number of test runs, then exploit the empirically best arm for the remaining attempts. With 10 000 total plays, increasing the number of test runs per arm improves decision reliability but reduces the time left for exploitation. The total returns (mean ± std) for test budgets of 10–3000 per arm show that small test sizes produce high variance and sometimes incorrect choices, while extremely large test budgets waste plays on exploration. **The optimal balance occurs around 100–300 test runs per arm**, where the average cumulative reward is highest (~5800–5900). At that point, the strategy consistently identifies the best-performing bandit (usually A) and achieves near-maximum total reward.

#### Now realize the same exploration for the epsilon-greedy policy with several epsilon values

```python
# ε-greedy strategy
def policy_eps_greedy(cas: Casino, total_attempts: int, eps: float) -> int:
    names = cas.names
    counts = {k:0 for k in names}
    means = {k:0.0 for k in names}
    total_reward = 0
    t=0
    # Warm-start: pull each arm once.
    for k in names:
        r = cas.pull(k); counts[k]+=1; means[k]=r; total_reward+=r; t+=1
        if t>=total_attempts:
          return total_reward
    while t<total_attempts:
        # With probability ε, choose a random arm (explore).
        # Otherwise, choose the arm with highest current mean reward (exploit).
        if (rng.random()<eps):
          k = names[rng.integers(len(names))]
        else:
          k = max(names, key=lambda x: means[x])
        r = cas.pull(k)
        total_reward+=r
        counts[k]+=1
        # mean update formula
        means[k]+= (r-means[k])/counts[k]
        t+=1
    return total_reward

# Each strategy will be run for 10 000 pulls per simulation.
total_attempts = 10_000
# ------------------------------Epsilon-greedy----------------------------------
# test different ε values (exploration probabilities).
# Report average total reward after 10 000 pulls.
eps_grid = [0.01, 0.05, 0.1, 0.2, 0.3, 0.5]
df_eps = pd.DataFrame([
    {"epsilon": eps, "Total attempts": total_attempts, **dict(zip(["Return mean (5 trials)","Return std"],
    simulate_policy(lambda cas, T: policy_eps_greedy(cas, T, eps), total_attempts, n_trials=5)))} for eps in eps_grid
])
display_dataframe_to_user("Epsilon-greedy — returns vs epsilon (mean±std over 5 trials)", df_eps)
```

The ε-greedy strategy balances exploration and exploitation by selecting a random arm with probability ε and the best-known arm otherwise. Simulations with 10 000 total plays show that small ε values (0.01–0.05) lead to premature exploitation and sub-optimal rewards, while very large ε values (≥ 0.3) over-explore and waste attempts. **The optimal trade-off occurs around ε = 0.1**, yielding the highest mean cumulative reward (≈ 5900) with moderate dispersion. After a few hundred plays, the policy consistently converges to the best bandit (A), indicating sufficient exploration to identify the most rewarding option.

#### And finally UCB with several values of parameter beta.

```python
# Upper Confidence Bound strategy
def policy_ucb(cas: Casino, total_attempts: int, beta: float) -> int:
    names = cas.names
    counts = {k:0 for k in names}
    means={k:0.0 for k in names}
    total_reward=0
    t=0
    for k in names:
        r = cas.pull(k)
        counts[k]+=1
        means[k]=r
        total_reward+=r
        t+=1
        if t>=total_attempts:
          return total_reward
    while t<total_attempts:
        t_eff = sum(counts.values())
        # pull each arm once
        # compute its upper confidence bound
        # Choose the arm with the largest UCB value.
        # Pull that arm, update mean and counts.
        ucb_vals = {k: means[k] + np.sqrt(beta * np.log(max(t_eff,1)) / counts[k]) for k in names}
        k = max(names, key=lambda x: ucb_vals[x])
        r = cas.pull(k)
        total_reward+=r
        counts[k]+=1
        means[k]+= (r-means[k])/counts[k]
        t+=1
    return total_reward

# Each strategy will be run for 10 000 pulls per simulation.
total_attempts = 10_000
# -----------------------------------UCB----------------------------------------
# Tests different UCB exploration parameters β.
# Collects mean & std total rewards into a DataFrame.
beta_grid = [0.5, 1.0, 2.0, 4.0, 5.0, 6.0]
df_ucb = pd.DataFrame([
    {"beta": beta, "Total attempts": total_attempts, **dict(zip(["Return mean (5 trials)","Return std"],
    simulate_policy(lambda cas, T: policy_ucb(cas, T, beta), total_attempts, n_trials=5)))} for beta in beta_grid
])
display_dataframe_to_user("UCB — returns vs beta (mean±std over 5 trials)", df_ucb)
```

The Upper Confidence Bound (UCB) algorithm selects arms by combining estimated mean rewards with a confidence term proportional to $(\sqrt{\frac{\beta \ln t}{n_i}})$. Increasing β encourages more exploration. Simulations with 10 000 total plays show that moderate **β values (0.5–1.0) yield the highest mean cumulative reward (~5900)** with relatively low dispersion. For larger β values (≥ 4), excessive exploration reduces overall performance. Thus, UCB efficiently learns the optimal arm and achieves strong performance even with limited parameter tuning. The best-performing bandit (A) becomes evident within the early phase of play as the uncertainty in estimates diminishes.

#### Disucss the relative merits and drawbacks for the A/B testing, epsilon-greedy, and USB.

### **1. A/B Testing**

Merits:

* Simple and intuitive to implement.
* Useful when exploration budget and total trials are fixed.
* Gives a clear statistical comparison of different options (common in experiments and website optimization).

Drawbacks:

* Rigid separation between exploration and exploitation — once the best arm is chosen, it never adapts if early data was misleading.
* Wastes trials during exploration on clearly inferior arms.
* Slower adaptation to noisy environments.

In this experiment:

* Performed reasonably for moderate test runs per arm (≈100–300),
  but too few runs caused misidentification, and too many wasted trials on exploration.



### **2. ε-Greedy Policy**

Merits:

* Simple and adaptive that exploration continues indefinitely at a controlled rate.
* Performs well with an optimal ɛ (~0.1 in this case).
* Easy to tune and understand.

Drawbacks:

* Uses uniform exploration that even bad arms are tried randomly.
* Requires manual tuning of ε that too small value results in premature convergence and too large value operates too much random play.
* Does not consider uncertainty explicitly that exploration is purely random.

In this experiment:

* ε ≈ 0.1 gave the best mean reward (~5900), balancing learning and exploitation effectively.


### **3. UCB (Upper Confidence Bound)**

Merits:**

* Mathematically grounded that exploration is driven by uncertainty, not random choice.
* Adapts automatically since explores new arms less as confidence increases.
* Very efficient that typically achieves near-optimal reward with fewer wasted trials.

Drawbacks:

* Requires careful choice of $\beta$ (exploration parameter). Too high value results in over-exploration and too low value does under-exploration.
* Slightly more complex to implement.

In this experiment:

* Moderate β (≈ 0.5–1.0) achieved the highest return (~5910), outperforming random exploration.
* For large β ≥ 4, excessive exploration reduced performance.
